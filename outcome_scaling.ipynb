```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("preprocess_2.csv", low_memory=False)
```

```python
df.head()
```

```python
y = df[['Install per Avg','Rating3']]
```

```python
scaler = StandardScaler()

# Fit the StandardScaler
y_scaler = scaler.fit(y)
y_scaled = y_scaler.transform(y)
```

```python
# pd.DataFrame(pca_x, columns=['pc1','pc2'])
# combined = df.join(df_y, how='inner')

y_df = pd.DataFrame(y_scaled, columns=['install_scaled','rating_scaled'])

df_merge = df.join(y_df, how='inner')
df_merge = df_merge.dropna()
df_merge
```

```python
df_merge.to_csv("preprocess_3.csv", index=False)
```
