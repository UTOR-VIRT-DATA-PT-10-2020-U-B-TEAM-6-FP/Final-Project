```python
import pandas as pd
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Drop additional features not needed for machine learning

ColumnsToDrop = [
    'fancyname',
    'company',
    'low_price_item',
    'top_price_item',
    'price',
    'game',
    'Released',
    'Last Updated',
    'appname',
    'numberreviews',
]
df_selectedColumns = df.drop(ColumnsToDrop,axis=1)
df_selectedColumns
```

```python
df_encoded = pd.get_dummies(df_selectedColumns, columns=['age_rating', 'categorygame','CompanyIsExperienced'])

df_encoded
```
