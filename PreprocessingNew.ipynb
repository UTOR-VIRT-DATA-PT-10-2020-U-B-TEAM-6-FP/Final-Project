```python
#Import file & resources

import pandas as pd
import numpy as np
```

```python
#Read file into df
df_accidents = pd.read_csv("Resources\Accidents0515.csv", low_memory=False)
```

```python
df_casualties = pd.read_csv("Resources\Casualties0515.csv", low_memory=False)
```

```python
df_vehicles = pd.read_csv("Resources\Vehicles0515.csv", low_memory=False)
```

```python
df_casualties
```

```python
df_casualties["Car_Passenger"].value_counts()
```

```python
#df_notna = df[df['numberreviews'].notna()]
df_accidents = df_accidents[df_accidents['Accident_Severity'] != -1]
```

```python
df_accidents
```

```python
df_vehicles
```

```python
df_accidents_withDriverData = pd.merge(df_accidents, df_casualties.loc[df_casualties["Casualty_Class"] == 1][["Accident_Index","Age_of_Casualty","Sex_of_Casualty","Casualty_Severity"]], how='inner', left_on='Accident_Index', right_on='Accident_Index')
```

```python
df_accidents_withDriverData
```

```python
df_accidents_withDriverData = df_accidents_withDriverData.rename(columns={'Age_of_Casualty': 'Age_of_Driver', 'Sex_of_Casualty': 'Sex_of_Driver', 'Casualty_Severity': 'Driver_Casualty_Severity'})
```

```python
Phase1Columns = ['Accident_Index',
'Accident_Severity',
'Number_of_Vehicles',
'Number_of_Casualties',
'Date',
'Day_of_Week',
'Time',
'Road_Type',
'Speed_limit',
'Light_Conditions',
'Weather_Conditions',
'Road_Surface_Conditions',
'Special_Conditions_at_Site',
'Carriageway_Hazards',
'Urban_or_Rural_Area',
'Age_of_Driver',
'Sex_of_Driver',
'Driver_Casualty_Severity']
df_accidents_withDriverData_Phase1 = df_accidents_withDriverData[Phase1Columns]
```
